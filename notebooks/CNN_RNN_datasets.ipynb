```python
def pad_one_image(img, target_size=(128, 128)):
    img_out = np.random.randint(0, 255, target_size)
    shape = img.shape
    xx = np.random.randint(0, 1 + target_size[0] - shape[0])
    yy = np.random.randint(0, 1 + target_size[1] - shape[1])
    img_out[xx : xx+shape[0], yy : yy+shape[1]] = img
    return img_out


# def pad_images(data, target_size=(128, 128)):
# #     data_out = np.random.randint(0, 255, (data.shape[0], target_size[0], target_size[1]))
#     data_len = data.shape[0]
#     data_out = np.empty((data_len, target_size[0], target_size[1]))
#     shape = data.shape[1:]
#     for ii in range(data_len):
# #         show(data_out[ii])
# #         show(data[ii])
#         data_out[ii] = pad_one_image(data[ii], target_size)
#     return data_out


def pad_images(data, target_size=(128, 128)):
    data_len = data.shape[0]
    data_out = np.random.randint(0, 255, (data_len, target_size[0], target_size[1]))
    shape = data.shape[1:]
    msg = "{} image will not fit in {} image"
    assert target_size >= shape, msg.format(shape, target_size)
    for ii in range(data_len):
        xx = np.random.randint(0, 1 + target_size[0] - shape[0])
        yy = np.random.randint(0, 1 + target_size[1] - shape[1])
        data_out[ii, xx : xx+shape[0], yy : yy+shape[1]] = data[ii]
    return data_out
```
